```python
import pandas as pd
movie = pd.read_csv('movies_dataset_prev.csv')
display(movie[['id', 'title', 'overview', 'genre']])
```

```python
movie['tags'] = movie['overview'] + movie['genre']
```

```python
display(movie[['id', 'title', 'overview', 'genre', 'tags']])
```

```python
new_data = movie.drop(columns=['overview', 'genre', 'original_language'])
```

```python
display(new_data[['id', 'title', 'tags']])
```

Convert Text - sklearn - vector

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
ct_v =CountVectorizer(max_features=10000, stop_words='english')
```

```python
ct_v
```

```python
vector = ct_v.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title'] == "The Godfather"][['id', 'title', 'tags']]
```

```python
new_data.info()
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]]['title'])
```

```python
def recommand (movie):
  index = new_data[new_data['title']== movie].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
  for i in distance[0:10]:
    print(new_data.iloc[i[0]]['title'])
```

```python
recommand ("Avatar")
```
